```python
import matplotlib.pyplot as plt
import networkx as nx
import random
```

```python
p=0.8 #the probability that a node is on
dim=20
G=nx.grid_2d_graph(dim, dim)
nx.draw(G)
plt.savefig("C:/Users/Narges/Documents/GitHub/computational_physics/chapter4/Fig/InitialGraph.png")
```

```python
for i in list(G.nodes):
    if (1-p)>random.random(): 
        G.remove_node(i)
        #print(i)
nx.draw(G)
plt.savefig("C:/Users/Narges/Documents/GitHub/computational_physics/chapter4/Fig/FinalGraph.png")
```

```python
check=False
for i in range(dim-1):
    for j in range(dim-1):
        if ((0,i) in G.nodes) and ((dim-1,j) in G.nodes):
            check=nx.has_path(G, (0,i), (dim-1, j))
if check==True:
    print("1")
if check==False:
    print("0")
```
